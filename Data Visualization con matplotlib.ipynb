Iniziamo a vedere alcuni semplici plot, disponibili anche in pandas:

```python
plt.plot([5,7,2,4]) # in ordinata
                    # una funzione (non un metodo)
```

```python
plt.plot([5,7,2,4], [4,6,9,2], 'ro') # due liste (per ascisse ed ordinate); 'ro' specifica il puntino (circle marker)
                                     # di colore red
```

```python
plt.plot([5,7,2,4], [4,6,9,2], 'bo') # due liste (per ascisse ed ordinate); 'ro' specifica il puntino (circle marker) 
                                     # di colore blue
```

```python
plt.plot([5,7,2,4], [4,6,9,2], 'r^') # possibilitÃ  come marker:
                                     # '.'          point marker
                                     # ','          pixel marker
                                     # 'o'          circle marker
                                     # '*'          star marker
                                     # '^'          triangle up marker
                                     # ecc (vedi help con shift+tab)
```

```python
plt.plot([5,7,2,4], [4,6,9,2], 'b.') # possibilitÃ  come colore:
                                     # 'b'          blue
                                     # 'g'          green
                                     # 'r'          red
                                     # 'c'          cyan
                                     # 'm'          magenta
                                     # 'y'          yellow
                                     # 'k'          black
                                     # 'w'          white

                                     # '.' fornisce uno scatterplot
```

```python
# creiamo due oggetti (utili dopo)
# sono i DATI:
x = [50, 70, 90, 65]
y = [129, 192, 163, 172]
```

Attenzione: i dati sono i dati (qui x e y). In un plot a linea, i dati sono UNITI da una linea, ma Ã¨ solo un artificio grafico.

```python
plt.plot(x, y, linewidth = 4.0) # linewidth specifica lo spessore della linea
```

```python
help(plt.plot) # alternativa per l'help
```

```python
plt.plot(x, y, linewidth = 2.0, linestyle = '--') # linestyle specifica lo stile della linea
                                                  # '-'          solid line style
                                                  # '--'         dashed line style
                                                  # '-.'         dash-dot line style
                                                  # ':'          dotted line style
                                                  # ecc (vedi help con shift+tab)
```

```python
plt.plot(x, y, linewidth = 1.0, ls = '-', marker = "o", markersize = 3) # marker come argomento a sÃ¨
```

```python
plt.plot(x, y, linewidth = 1.0, linestyle = '-', marker = "o", markersize = 10, markerfacecolor = 'red') 
                                                 # puntini vuoti (bianchi dentro) di dimensione 10
```

```python
plt.plot(x, y, lw = 1.0, ls = '-', marker = "o", ms = 10, mfc = 'white') # con alcuni argomenti (parole multiple?) abbreviati
```

Sinora abbiamo visto **singole funzioni** di plotting fornite da 'matplotlib'.<br>
Entriamo ora nel cuore di 'matplotlib', che infatti permette di **costruire il plot poco per volta** - non possibile con pandas che semplicemente fornisce dei metodi di plotting per le classi 'dataframe' e 'serie', con i relativi argomenti tra ().<br>


```python
plt.plot() # il grafico vuoto (solo le coordinate) - SENZA DATI (i valori dei tick degli assi?)
```

```python
# specifichiamo il titolo del grafico e il nome degli assi
```

```python
plt.plot(x, y)           # il grafico con i dati (e i tick corrispondenti)
```

```python
plt.plot(x, y)           # il grafico vuoto (solo le coordinate)
plt.title("TITLE")
plt.xlabel("Asse X")
plt.ylabel("Asse Y")
```

```python
# come la cella prima, con i colori in piÃ¹:
plt.plot(x, y, color = "red")
plt.title("TITLE", color = "blue")
plt.xlabel("Asse X", color = "purple")
plt.ylabel("Asse Y", color = "green")
```

```python
plt.plot(x, y)
plt.title("TITLE", color = "blue")
plt.xlabel("Asse X", color = "purple")
plt.ylabel("Asse Y", color = "green")
plt.grid(True)            # griglia
plt.legend(['Legend1'])   # legenda (loc = 1 def, angolo top-right)
```

```python
plt.plot(x, y)
plt.title("TITLE", color = "blue")
plt.xlabel("Asse X", color = "purple")
plt.ylabel("Asse Y", color = "green")
plt.grid(True)
plt.legend(['Legend1'], loc = 2) # legenda in posizione differente (angolo top-left)
```

Possibili posizioni della legenda:

* 3 in basso a destra
* 4 in basso a sinistra
* 5 a destra
* 6 centrato a sinistra
* 7 centrato a destra
* 8 centrato basso
* 9 centrato alto
* 10 al centro

Codici dei colori:

* b blu
* c ciano
* g verde
* m magenta
* r rosso
* y giallo
* k nero
* w bianco

```python
# 3 plot sovrapposti
plt.plot([1,2,3,4],[1,4,8,15],'b*')
plt.plot([1,3,5,7],[1,4,8,12],'g^')
plt.plot([1,2,3,5],[2,5,4,12],'ro')
plt.legend(['First','Second','Third'],loc=0)  # 3 legende, una per serie
```

```python
# creiamo piÃ¹ sottografici (equivalente di 'par=mfrow' di R).
# i numeri tra parentesi indicano numero di righe e di colonne del layout e numero del sottografico
```

```python
# plt.subplot (n1,n2,n3) dove n1 e n2 sono le posizioni della matrice dei plot ed n3 Ã¨ la posizione sulla quale lavorare:

plt.subplot(2,2,1) # sottografico 1 del layout 2x2
plt.plot([1,2,3,4],[1,4,8,15],'b*')

plt.subplot(2,2,2) # sottografico 2 del layout 2x2
plt.plot([1,3,5,7],[1,4,8,12],'g^')

plt.subplot(2,2,3) # sottografico 3 del layout 2x2
plt.plot([1,2,3,5],[2,5,4,12],'ro')

plt.subplot(2,2,4) # sottografico 4 del layout 2x2
plt.plot([1,2,3,5],[2,5,4,12],'b')
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[1,4,8,15],'b*')

plt.subplot(1,2,2)
plt.plot([1,3,5,7],[1,4,8,12],'g^')
```
