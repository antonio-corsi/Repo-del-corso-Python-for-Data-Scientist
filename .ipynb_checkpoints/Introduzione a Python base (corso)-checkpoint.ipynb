- Sottrazione: `-`

Un termine migliore di *variabile* è *oggetto* (appartenente ad una classe, visibile con la funzione `type`).

````python
# int
a = 100
print('a: ',a,type(a))

# float
b = 5.1
print('b: ',b,type(b))

# boolean
dv = False
print('dv: ',dv, type(dv))

# string
str = 'stringa'
print('str: ', str, type(str))

# list
list = [1,2,3,"salve",200.5]
print('list: ',list,type(list))

# dictionary (the LOGICAL structure of data, indipendente dal formato, sulla quale NON possiamo lavorare)
dict = {'lab':['A', 'B', 'C','D'], 'val':[10, 30, 20,40]}
print('dict: ',dict,type(dict))

# creiamo un oggetto della classe 'dataframe' (che utilizza 'dict'):

import pandas as pd
df = pd.DataFrame(dict)  # il metodo initiator (costruttore) che CREA l'oggetto di classe dataframe
                         # attenzione all'uso delle PARENTESI:
                         # - le () delimitano gli argomenti del metodo (o della funzione)
                         #   in Python, gli argomenti di una funzione o di un metodo sono chiamati PARAMETERS
                         # - le [] delimitano un elenco di valori oppure fanno subsetting
                         # - le {} delimitano il dictionary (la definizione di metadati)

print(type(df))
df
````

## Conversione tra Tipi

È possibile convertire un tipo in un altro usando le funzioni `int()`, `float()`, `str()`, `bool()`, ecc.

````python
stringa = "123"                              # una stringa
print(stringa, type(stringa))

# Conversione a intero
numero_intero = int(stringa)                 # trasforma la stringa in un intero
print(numero_intero, type(numero_intero))

# Conversione a floating point
numero_mobile = float(stringa)                 # trasforma la stringa in un intero
print(numero_mobile, type(numero_mobile))

# Conversione a stringa
# del(str)
stringa = str(numero_intero)                 # trasforma la stringa in un intero
print(stringa, type(stringa))
````

Il modo migliore per gestire i file è usare la parola chiave `with`. In questo modo si chiude automaticamente il file quando non serve più, cioè in questo modo non è necessario chiamare il metodo `.close()`.

Per esempio:

````python
with open("esempio_testo.txt", "r") as f:
    data = f.read()
    print(data)
````

````python
with open("esempio_testo.txt", "r") as f:
    data = f.readline()
    print(data)
````

È possibile scrivere in coda ad un file. L'argomento `r+` consente sia la lettura che la scrittura sul file.

````python
with open("esempio_testo.txt", "r+") as f:
    data = f.readline()
    f.write("Hello World!")
````

Vediamo come funzionnao i `set` e come manipolarli in vari modi:

I set sono un tipo di dati Python **immutabile, non ordinato e privo di elementi duplicati**.

````python
empty_set = {}
empty_set2 = set()
````

**Immutabile** significa che agli elementi dell'insieme non possono essere assegnati valori diversi. Ad esempio, se avessimo un insieme `z = {1, 2, 3, 4, 5}`, non potremmo riassegnare l'elemento 1 a 2.

**Non ordinato** significa che gli elementi dell'insieme non hanno un ordine.

**Nessun duplicato** significa che gli stessi elementi non possono comparire più di una volta nello stesso insieme.

Inseriamo ed eseguiamo il seguente codice per creare il nostro primo *set*:

````python
my_first_set = {1,2,3,4,5}
````

# Flussi di Controllo in Python

I flussi di controllo permettono di eseguire blocchi di codice in base a condizioni (`if`) o di ripetere blocchi più volte (`for`, `while`).

## Condizioni: if, elif, else

La struttura condizionale permette di eseguire codice solo se una certa condizione è vera.

````python
x = 10
# x = -10

if x > 0:
    print("x è positivo")     # no else branch
````

Il segno `:` determina l'inizio di un blocco (associato a quel ramo della if).
L'indentazione determina lo scope del blocco.

````python
x = 10
# x = -10

if x > 0:
    print("x è positivo")
    # maybe some other instructions
else:                         # the else branch (again with the colon)
    print("x è negativo")
````

````python
x = 10
# x = -10
# x = 0

if x > 0:
    print("x è positivo")
elif x == 0:                  # una if annidata (nested)
    print("x è zero")
else:
    print("x è negativo")
````

## Cicli for

Il ciclo `for` permette di **iterare** su una sequenza (lista, stringa, range, ecc.).<br>
Non confondere `frutti`, la sequenza sulla quale iteriamo, con `frutto`, l'istanza corrente (la variabile di controllo).<br>
Come per le if, si usa il segno `:` per iniziare un blocco, il cui scope è definito dalla indentazione.<br>
Le due parole chiave qui sono `for ` a `in`.

````python
# Iterazione su una lista
frutti = ["mela", "banana", "ciliegia",10]
for frutto in frutti:          # la variabile di controllo NON deve essere definita
    print(frutto)

# Iterazione su un intervallo di numeri
for i in range(5):
    print("Valore:", i)
````

**Nota sul `range`, sulla `print` e sul *piping***

Vediamo qualche esempio di `range`

````python
range(10)          # il limite superiore (upper limit) NON è incluso in the range.
                   # python conta da 0, perciò il primo elemento è 0.
                   # il parametro della funzione range indica la lunghezza del range (e non il  suolimite superiore)
````

````python
print(range(10))   # il limite superiore (upper limit) NON è incluso in the range.
                   # python conta da 0, perciò il primo elemento è 0.
                   # il parametro della funzione range indica la lunghezza del range (e non il  suolimite superiore)

type(range(10))    # classe 'range'
````

Ogni cella markdown visualizza automaticamente l'ultimo output. Se vogliamo visualizzare tutti gli output della cella, dobbiamo usare la funzione `print` **per ogni output** della cella.

Naturalmente possiamo usare la funzione `print` anche per l'ultimo output, come nel seguente codice:

````python
print(range(10))          # il limite superiore (upper limit) NON è incluso in the range.
                          # python conta da 0, perciò il primo elemento è 0.
                          # il parametro della funzione range indica la lunghezza del range (e non il  suolimite superiore)

print(type(range(10)))    # classe 'range'
````

Come si vede il formato dell'ultimo output è differente, ma l'informazione in esso contenuta è la stessa.

> L'ultima riga della cella precedente (`print(type(range(10)))`) è un ottimo esempio di **piping**. L'output della prima funzione (quella più interna, cioè `range(10)`) è passato come input alla seconda funzione (`type`), e così via. ## Cicli while

Il ciclo `while` ripete un blocco finché una **condizione è vera**.

````python
i = 0
while i < 5:
    print("i vale:", i)
    i += 1                # questo è un semplice e compatto modo in Python per scrivere 'i = i+1'
````

## Parole chiave `break` e `continue`

- `break` interrompe il ciclo.
- `continue` salta all'iterazione successiva.

````python
for i in range(10):                     # 'i' è la variabile di controllo del ciclo for, 'range(10)' è l'oggetto sul quale vogliamo iterare
    if i == 5:                          # '==' in poython è usato per verificare una condizione, '=' è usato per le assegnaizoni di valori
        break                           # ESCE dal loop
    print("Valore con break:", i)

for i in range(10):
    if i % 2 == 0:                      # i è PARI
        continue                        # salta un'iterazione
    print("Valore con continue:", i)    # --> otteniamo la stampa solo dei valori di i DISPARI
````

# Il significato del simbolo `:`

Il **simbolo :** in Python è uno dei caratteri più importanti della sintassi. Serve per iniziare **un blocco di codice indentato**, e lo trovi in tantissimi contesti. Ecco una panoramica dei principali:

1. **Dopo le strutture di controllo** (if, for, while, ecc.)
```python

    print("Positivo")
    
for i in range(5):
    print(i)

while x < 10:
    x += 1
```
    Dopo il : segue un blocco indentato che definisce cosa fare in quel caso.

2. Nelle **definizione di funzioni e metodi delle classi**
   
```python

    print(f"Ciao {nome}")
    
class Persona:
    def __init__(self, nome):
        self.nome = nome
```
    Anche qui il : segnala che inizia il corpo della funzione o classe.

3. Con **slicing (fette) di liste, stringhe, array**
   
```python
lista = [10, 20, 30, 40, 50]
print(lista[1:4])  # Output: [20, 30, 40]
```  
    Qui : indica un intervallo: start:stop (incluso start, escluso stop).<br>
    Altri esempi:
    
```python
s[:3]    # primi 3 elementi
s[2:]    # dal terzo elemento in poi
s[:]     # tutta la sequenza (copia)
```

*Le *dict comprehension* sono un modo compatto e leggibile per creare un dizionario in Python, <u>direttamente in una sola riga</u>, partendo da un iterabile (come una lista, una tupla, un'altra dict, ecc.). È l’equivalente per i dizionari di quello che sono le list comprehension per le liste.*


5. Nella dichiarazione di tipi con **type hinting** (dal Python 3.5+)<br>
*Il type hinting è una funzionalità introdotta in Python 3.5 (e potenziata nelle versioni successive) che permette di indicare i tipi di variabili, parametri e valori di ritorno delle funzioni. Non cambia il comportamento del codice, ma aiuta gli editor (come VSCode), i tool statici (come mypy) e anche i colleghi umani a capire meglio cosa aspettarsi.*

```python
def somma(a: int, b: int) -> int:
```

    Il : qui serve a specificare il tipo del parametro.

# Gli iteratori
Argomento non previsto a programma ma molto utile.<br>
Dal *go deploy lab*.

Un oggetto iteratore è qualcosa su cui è possibile iterare. Ciò significa che è possibile attraversare tutti gli elementi.

Un oggetto iteratore utilizza i [**metodi *dunder***](https://www.programmareinpython.it/video-corso-python-programmazione-a-oggetti/06-i-metodi-magici-dunder-methods/) `iter` e `next`.

Un **iterator** è un oggetto che contiene una funzione `next`, che restituisce l'elemento successivo dell'iteratore e lo interrompe quando non ci sono più elementi disponibili.

Gli iteratori non calcolano ogni elemento quando è istanziato, ma li calcolano solo uno alla volta quando vengono richiesti. Questo può essere molto più utile di altre tecniche in determinati scenari.

Ora creeremo un semplice esempio di iterator. Inserisci ed esegui il seguente codice.

````python
# definizione della classe 'MyIterator' con 2 metodi (iter e next)

class MyIterator():                       # una classe definita dall'utente ("user-defined class")

    # il metodo costruttore
    def __iter__(self):                   # 'self' è l'istanza della classe, ed è valorizzato alla chiamata del metodo
        self.value = 10                   # assegna 10 all'istanza.
                                          # 'self' è  la memoria interna dell'istanza
        return self                       # l'istanziazione della classe, cioè la creazione dell'oggetto appartenente a quella classe, restituisce l'oggetto 'self'.

    def __next__(self):
        new_value = self.value * 10
        self.value = new_value
        if new_value > 1000000:
            raise StopIteration
        return new_value

# l'istanziazione della class (nel gergo della OOP si dice che 'si istanzia la classe', NON che la si chiama.
# è un processo a due step (perchè?)
my_class = MyIterator()   # in questo modo creiamo una NUOVA ISTANZA della classe (è un oggetto), qui di nome 'my_class' (una struttura vuota).
iterator = iter(my_class) # creazione dell'iteratore (una sintassi OOP differente dalla quella usuale: oggetto.method (cioè: my_class.iter)
                          # il 'value' dell'istanza creata è 10.

# prova:
x = True
while x:
    print(next(iterator))
````

Esaminiamo cosa abbiamo fatto nella cella precedente.

Abbiamo creato una classe chiamata `MyIterator()`, quindi aggiungiamo i due metodi *dunder* `iter` e `next` che la definiscono e la usano come un oggetto iterator.

Impostiamo un valore iniziale pari a 10.

Quindi il metodo dunder `next` moltiplica quel valore per dieci e aggiungiamo anche una condizione in modo che, se il valore supera 1 milione, il metodo next interrompa il programma.

Quindi creiamo un semplice ciclo per richiamare continuamente il metodo `next`. Osserviamo che l'output è 10, 100, 1000, ecc... Sopra 1000000 il ciclo si interrompe.

# I decoratori
Dal *go deploy lab*.

I decoratori Python servono ad estendere le funzionalità delle funzioni.

I decoratori sono funzioni che accettano altre funzioni come input. In Python anche le funzioni sono trattate come oggetti.

I decoratori sono utili per estendere la funzionalità di determinate funzioni senza modificare effettivamente la funzione originale. Questa tecnica può offrire una gamma incredibilmente ampia di vantaggi.

Inizieremo creando un decoratore molto semplice.

````python
def function_1(func): # function that has another function as a parameter
    def wrapper(): # wrapper function that adds the functionality
        print("Prior") # code that acts prior to function being called
        func() # calling the original function
        print("Post")   # code that acts after the function being called
    return wrapper # returns the wrapper function
````

I decoratori hanno questo semplice formato.

Ora creeremo una semplice funzione `Hello world` e vi applicheremo il decoratore. Il modo per applicare una funzione decoratore a un'altra funzione è inserire `@decorator_name` immediatamente sopra la funzione che si desidera decorare. Per esempio:

````python
def function_1(func):
    def wrapper():
        print("Prior")
        func()
        print("Post")
    return wrapper


@function_1
def function_2():
    print("Hello World!")

function_2()
````

Quindi, quando si esegue questo codice, invece di ottenere semplicemente *Hello World!*, il decoratore aggiungerà *Prior* e *Post* all'output.

Ora aggiungeremo una seconda funzione e la decoreremo. Si noti come questa tecnica permetta di aggiungere ulteriori funzionalità a qualsiasi funzione che scegliamo, in modo incredibilmente semplice e veloce.

````python
def function_1(func):
    def wrapper():
        print("Prior")
        func()
        print("Post")
    return wrapper


@function_1
def function_2():
    print("Hello World!")

@function_1
def function_3():
    print("Goodbye")

function_2()
function_3()
````

Ora creiamo un esempio pratico e utile di decoratore. Per prima cosa, creiamo un decoratore chiamato `@timer_decorator`, che cronometra il tempo di esecuzione di un'altra funzione. Dopodiché, potremo semplicemente chiamare `@timer_decorator` sopra qualsiasi funzione che vogliamo cronometrare.

**NON ESEGUIRE la cella seguente ---> va in loop (infinito?)**

````python
import time

# this decorator takes a function and times it, then prints the time it took to execute
def timer_decorator(func):
    def wrapper():
        start_time = time.perf_counter()
        func()
        end_time = time.perf_counter()
        timer = end_time - start_time
        print(f"Time for {str(func.__name__)}: " + str(timer))
    return wrapper

# this decorator adds a 1 second delay to a function
def slow_down(func):
    def wrapper():
        time.sleep(1)
        func()
        print("Function delayed by 1 second")
    return wrapper

# here we are decorating a simple nested function to time it
@timer_decorator
def function_4():
    i = 0
    max = 1000
    mylist = []
    while i < max:
        j = 0
        while j < max:
            k = i * j
            mylist.append(k)
        j += 1
    i += 1
    return len(mylist)

# here we are decorating a function twice to slow it down and decorate it
@timer_decorator
@slow_down
def function_5():
    i = 0
    max = 100
    mylist = []
    while i < max:
        j = 0
        while j < max:
            k = 0
            while k < max:
                z = i * j * k
                mylist.append(z)
                k += 1
            j += 1
        i += 1
    return len(mylist)

function_4()
function_5()
````

# Le funzioni ricorsive
Dal *go deploy lab*.

Le funzioni ricorsive sono **funzioni che chiamano se stesse**. Sappiamo che le normali funzioni Python possono chiamare altre funzioni, ma ora esamineremo le funzioni che chiamano se stesse.

Ne creeremo una. Questa funzione inizierà da `x = 0`, poi lo incrementerà di uno, quindi la funzione stamperà `x` e poi chiamerà se stessa, questa poi riavvierà la funzione e incrementerà di uno e stamperà di nuovo, e così viasino a 10.

````python
x = 0                                    # assegnazione del valore del parametro effettivo ("actual" in english)

# la definizione della funzione:
def recursive_function_example(x):       # nuovamente è una funzione definita dall'utente
    x += 1                               # --> x = x + 1
    print(x)
    if x == 10:
        return print("Stop")             # stop della funzione, che rewstituisce 'print("Stop")'
    recursive_function_example(x)        #--> la RICORSIONE!

# la chiamata della funzione
recursive_function_example(x)            # qui il nome del parametro effettivo coincide con quello del parametro formale, ma non è necessario
````

Come si vede, abbiamo aggiunto una condizione di arresto con l'istruzione if x == 10. Se non avessimo aggiunto questa condizione di arresto, la funzione sarebbe stata eseguita all'infinito.

Un esempio comune di funzione ricorsiva è la [**sequenza di Fibonacci**](https://it.wikipedia.org/wiki/Successione_di_Fibonacci). A questo scopo inseriamo ed eseguiamo il seguente codice:

````python
# il solito processo a 3 step.

# 1. assegnazione dei valori ai parametri effettivi della funzione
x = 0
y = 1

# 2. definizione della funzione
def fibonacci(x,y):
    print(x)
    z = x + y
    x = y
    y = z
    if x > 1000:
        return print("Stop")
    fibonacci(x,y) # calls itself

# 3. chiamata della funzione
fibonacci(x,y)
````

````python
# Normal
my_list = [1,2,3,4,5]
new_list = []
for x in my_list:
    x = x**2
    new_list.append(x)
print(new_list)

# List comprehension
new_list2 = [x**2 for x in my_list]
print(new_list2)
````

Il controllo dei tipi è forte (strong typing) e viene eseguito in runtime (dynamic typing): una variabile è un contenitore a cui viene associata un'etichetta (il nome) che può essere associata a diversi contenitori anche di tipo diverso durante il suo tempo di vita. # Object-oriented programming (OOP) in Python

**Nota sulla OOP**:
* un linguaggio ad oggetti non fornisce (solo) delle *funzioni* ma fornisce delle *classi* con dei *metodi*
* una classe è una collezione di oggetti (istanze della classe)
* i metodi della classe sono gli unici abilitati a lavorare sugli oggetti - è il concetto dell'[incapsulamento](https://it.wikipedia.org/wiki/Incapsulamento_(informatica))
* ogni oggetto (di una certa classe) è istanziato (cioè creato) tramite un metodo particolare detto 'costruttore' (in Python si chiama 'initiator', in altri linguaggi 'new').

In generale, *metodi* e *funzioni* in python sono cose diverse: per semplicità diciamo che un metodo è una funzione che è
strettamente collegata con un oggetto di una classe (ad es. un dataframe pandas, oppure una lista) e che ha accesso ai
suoi dati; una funzione (ad esempio, 'print' oppure 'type'), deve invece avere specificati i dati sui quali agire tramite i suoi argomenti in parentesi.<br>

Attenzione: sia i metodi delle classi che le funzioni usano le () per specificare gli argomenti in input.<br>

**Esempi tipici di classe**:

Il singolo cliente Mario Rossi è "creato" (cioè istanziato) tramite il metodo costruttore della classe Clienti. A questo punto questo nuovo oggetto (o istanza) Mario Rossi "eredita" tutti i metodi della classe "Cliente" (sopra elencati). L'oggetto Cliente è detto in genere un **business object**. Altri esempi tipici di oggetti di business sono: ordine, fattura, prodotto, magazzino, conto corrente, bancomat, ecc. In genere le classi di business sono definite dall'analista o dallo **sviluppatore**.

Consider that all instance-level attributes of a class must always be positive.

But the biggest issue is that these methods scale with the total attributes.

Imagine how this class implementation would look like with 10 such methods.

There's so much redundancy!

Also, validation checks defined in the setters must be defined in the __init__ method as well. This is done to ensure proper initialization.

More redundancy!

Python Descriptors is one of the finest solutions here.

Descriptors are objects with methods (like __get__, __set__, etc.) used to manage access and validate attributes of another class.

So, one descriptor object is assigned to only one attribute of another class.

A typical Descriptor class is implemented with just three methods (shown in the code below):

1) The __set_name__ method: Invoked when the descriptor is assigned to an attribute.
2) The __set__ method: Invoked when the attribute is assigned a new value. We can define the custom checks here.
3) The __get__ method: Invoked when the attribute is accessed.

Done!

This addresses both issues we discussed earlier:
↳ It reduces the redundancy as the same class implementation is used across attributes.
↳ The init stage can also ensure proper initialization, without explicit checks.

Moreover, the class implementation is much more cleaner because it's no longer overloaded with getters and setters.

I find descriptors to be massively helpful in reducing code redundancy and making the entire implementation much more elegant.

## Le classi

Ecco una breve **introduzione alla programmazione orientata agli oggetti (OOP)**. L'OOP è uno dei paradigmi di programmazione, ed è organizzato intorno ad **oggetti anzichè funzioni e logica**.

I **metodi** sono spesso definiti all'interno di **classi**. Un metodo è come una funzione che si trova <u>all'interno di una classe</u>.

Le classi sono strutture dati definite dallo sviluppatore (l'utente). Si possono poi creare **istanze di una classe** che utilizzano l'archetipo della classe. Ad esempio, si può creare una **classe Persona()**, che definisce le <u>caratteristiche</u> (anche dette "proprietà", "attributi", "parametri" o "variabili") che una persona ha, come altezza, peso, nome ed età. Quindi, al momento della istanziazione della classe, si possono assegnare dei valori a queste caratteristiche.

Una classe è un **blueprint** per la creazione di un'istanza. Un'istanza contiene <u>dati reali su qualcosa</u>.

Creiamo un oggetto persona che abbia un nome, un'età, ecc.

Ora creeremo una classe molto semplice; inseriamo e eseguiamo il seguente codice:

````python
class Person():
    pass
````

La parola chiave `class` è utilizzata per creare una classe. Gli si aggiunge il nome della classe in maiuscolo della (per convenzione, non necessariamente).

Il metodo `__init__()` definisce tutte le caratteristiche che TUTTE le istanze devono avere. `__init__()` inizializza ogni nuova istanza della classe. A questa funzione possiamo passare più parametri, ma `self` deve sempre essere il primo. Il parametro `self` passa l'istanza alla funzione `__init__()` in modo che le caratteristiche possano essere definite su di essa. Per esempio:

````python
class Person():
    def __init__(self,name, age, height, weight):    # la parola chiave 'def' definisce un metodo e NON una funzione perchè  qui siamo all'interno della definizione di una classe
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight
````

La riga di codice `self.x = x` serve per assegnare il valore `x` alla caratteristica `x`.

La classe `Person` afferma genericamente che tutte le "Person" hanno le caratteristiche nome, età, altezza e peso; poi si crea un'istanza di questa classe per definire un individuo reale con i suoi nome, età, altezza e peso.

Creiamo un'istanza di questa classe. Per esempio:

````python
class Person():
    def __init__(self,name, age, height, weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight

person_1 = Person("John", 40, 168, 180)   # l'istanziazione della classe
````

La cella precedente ha creato un oggetto `person_1` con caratteristiche John, 40 anni, alto 168 cm e pesante 180 libbre.

Tutti i parametri all'interno del metodo `__init__` (come `self.name`) hanno valori specifici a seconda dell'istanza. Al contrario, se assegniamo una caratteristica al di fuori del metodo `__init__`, essa diventa una caratteristica di classe, cioè  lo stesso valore per tutte le istanze. Per esempio:

````python
class Person():
        nation = "Italian"                            # una caratteristica di classe che si applica atutte le istanze.
                                                      # NON possiamo MODIFICARE questo attributo per ogni istanza (oggetto)
        def __init__(self,name, age, height, weight):
            self.name = name                          # caratteristica della istanza
            self.age = age                            # caratteristica della istanza
            self.height = height                      # caratteristica della istanza
            self.weight = weight                      # caratteristica della istanza

person_1 = Person("John", 40, 168, 180)
````

Accedere alle caratteristiche di un'istanza è molto semplice. Ad esempio, se vogliamo accedere al (cioè leggere il) nome di `person_1`, facciamo così:

````python
person_1.name                                         # estrae (legge) il nome di 'person_1'
````

Allo stesso modo:

````python
print(person_1.nation)
print(person_1.age)
print(person_1.height)
person_1.weight
````

Se vogliamo riassegnare i valori di queste caratteristiche di `person_1` facciamo semplicemente così:

````python
person_1.age = 70
person_1.age
````

## I metodi

### Metodi di istanza

I metodi sono funzioni che **appartengono a un oggetto**, cioè sono membri di una classe. Accettano un'istanza di classe come primo parametro.

Esistono **tre tipi di metodi**: metodi di classe, metodi di istanza e metodi statici. Analizzeremo ciascuno di questi.

I metodi più comuni sono i metodi di istanza.

I metodi di istanza accettano un parametro predefinito, ovvero il parametro `self`, che è un'istanza della classe. Una funzione che sia creata all'interno della classe è di default un metodo di istanza.

Creiamo un semplice metodo di istanza (`say_hello`), aggiungendolo alla classe `Person` ed al metodo `__init__`.<br>
Quindi istanziamo la persona e invochiamo il nuovo metodo:

````python
class Person():
    nation = "Italian"

    def __init__(self,name, age, height, weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight

    def say_hello(self):
        print("Hello world!")

person_1 = Person("John", 40, 168, 180)
person_1.say_hello()
````

Non è necessario inserire il parametro `self` quando si chiama il metodo, Python lo fa automaticamente per noi.

Un altro esempio molto semplice è il seguente codice:

````python
class Summation():
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def addition(self):
        return self.x + self.y

z = Summation(5,6)
z.addition()
````

### I metodi di classe

I metodi di classe non possono accedere alle singole istanze. Sono associati alla classe e non all'istanza.

Utilizziamo un decoratore per specificare che il metodo è un metodo di classe.

Torniamo al primo esempio (`Person)` e definiamo un nuovo metodo di classe. Usiamo un decoratore `@classmethod` per definire il nostro metodo come metodo di classe. Scriviamo un metodo normale, ma anzichè inserire `self` come parametro inseriamo `cls` (per classe, lo facciamo solo per convenzione). Quindi, per invocare il metodo di classe, possiamo fare riferimento alternativamente all'oggetto o alla classe.

````python
class Person():
    nation = "Italian"
    def __init__(self,name, age, height, weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight

    def say_hello(self):
        print("Hello world!")

    @classmethod
    def what_nation(cls):        # 'cls' sta per CLASSE
        print(Person.nation)

person_1 = Person("John", 40, 168, 180)

# 2 sintassi alternative:
print(person_1.what_nation())
print(Person.what_nation)
````

### Metodi statici

I metodi statici non accettano parametri specifici come `cls` o `self`. Un metodo statico non richiede la creazione di un'istanza della classe. È indipendente da qualsiasi istanza ed è collegato alla classe stessa. Per definire un metodo come statico si utilizza il decoratore `@staticmethod`.

## Ereditarietà (*inheritance*)

Una classe può **ereditare** le caratteristiche e gli attributi di un'altra classe. Questa è l'ereditarietà.

Abbiamo una **classe madre** che è la classe originale, e una **classe figlia** che eredita le caratteristiche dalla classe madre.

Definiamo la classe con cui lavoreremo:

````python
class Person():
    nation = "Italian"
    def __init__(self,name, age, height, weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight
    def display_name(self):
        print(self.name)

person_1 = Person("John", 40, 168, 180)
person_1.display_name()
````

La classe `Person` che abbiamo creato è la classe madre. Ora creiamo una classe figlia che erediti le caratteristiche e i metodi della classe `Person`.

````python
class Employee(Person):
        pass

person_2 = Employee("Michael", 90, 102,291)
print(person_2.age)
person_2.display_name()
````

L'aggiunta di `pass` alla classe significa semplicemente che questa nuova classe `Employee` figlia è vuota. Cioè la classe `Employee` eredita semplicemente tutto dalla classe `Person()` e nient'altro.

Quindi, quando creiamo un'istanza della classe `Employee()`, possiamo accedere alla caratteristica `age` nello stesso modo di prima. Possiamo anche accedere al metodo `display_name()` come prima. E così via.

La classe `Employee` ha dunque (ereditato) tutti i metodi e le caratteristiche della classe Person. Se vogliamo aggiungere ulteriori metodi o caratteristiche lo possiamo fare.

Definiamo un metodo aggiuntivo alla classe `Employeee` (`weight_times_height`) e invochiamolo:

````python
class Employee(Person):
    def weight_times_height(self):            # moltiplica il peso per l'altezza
      return int(self.weight * self.height)

person_3 = Employee("Sarah", 74, 122,120)
print(person_3.weight_times_height())
````

Verifichiamo che il nuovo metodo `weight_times_height`non è disponibile alle istanze della classe madre.<br>
Come sappiamo da prima, `person_1` è stata istanziata dalla classe `Person` (e non dalla classe `Employee`):

````python
person_1.weight_times_height()
````

I metodi e le caratteristiche della classe figlia, se omonimi, sovrascrivono la classe padre. Ad esempio:

**Incapsulamento vs Information Hiding**<br>

---

- 1. **Incapsulamento (Encapsulation)**

* **Definizione**: È il concetto di **raggruppare** dati (attributi) e funzioni (metodi) che operano su quei dati all’interno di **un’unica entità** – di solito una classe.
* **Idea principale**: Mettere insieme comportamento e stato per rappresentare oggetti del mondo reale.
* **Obiettivo principale**: **Organizzare** il codice in modo modulare e riutilizzabile.

Esempio:

```python
class Auto:
    def __init__(self, velocità):
        self.velocità = velocità   # stato
    def accelera(self):           # comportamento
        self.velocità += 5
```

Tutto ciò che riguarda un’`Auto` è incapsulato nella classe `Auto`.

---

2. **Nascondimento dell’informazione (Information Hiding)**

* **Definizione**: È la pratica di **limitare l’accesso** ai dettagli interni di una classe o modulo che **non servono** all’utente esterno.
* **Idea principale**: Nascondere **come** funziona qualcosa, mostrando solo **ciò che serve**.
* **Obiettivo principale**: **Proteggere** lo stato interno e imporre regole d’uso corrette.

Esempio:

```python
class ContoBancario:
    def __init__(self):
        self.__saldo = 0   # nascosto (con name mangling)

    def deposita(self, importo):
        if importo > 0:
            self.__saldo += importo

    def saldo_corrente(self):
        return self.__saldo
```

In questo caso, `__saldo` è nascosto: chi usa la classe non può modificarlo direttamente, ma solo attraverso metodi sicuri.

---

Differenze principali

| Concetto                      | Incapsulamento                          | Nascondimento dell’informazione                     |
| ----------------------------- | --------------------------------------- | --------------------------------------------------- |
| **Scopo**                     | Raggruppare dati e metodi               | Nascondere i dettagli di implementazione            |
| **Focus**                     | Struttura e organizzazione              | Controllo degli accessi                             |
| **Si realizza tramite**       | Classi e oggetti                        | Modificatori d’accesso (es. `private`, `protected`) |
| **Preoccupazione principale** | Architettura del codice                 | Sicurezza e integrità dello stato interno           |
| **Visibilità**                | Non nasconde necessariamente i dettagli | Nasconde intenzionalmente ciò che è interno         |

---

### ✅ In breve:

* **Incamiciamento** = *“Metti tutto ciò che è correlato nello stesso contenitore.”*
* **Nascondimento dell’informazione** = *“Non farmi vedere più di quanto mi serve o mi è permesso toccare.”*

I due concetti **si completano** a vicenda: l’incapsulamento riguarda **la struttura**, il nascondimento dell’informazione riguarda **la protezione**.

# Pile (*stack*) e code (*queue*)

In questo capitolo vediamo le strutture dati **stack** e **queue**.

*Heap*, *stack* e *queue* sono differenti tipi di strutture dati che impareremo a implementare con Python.

Una pila (*stack*) è una struttura dati in cui l'ultimo elemento inserito è il primo ad essere rimosso. Immagina una pila di libri o di pratiche di un ufficio: è come mettere un libro su una pila e poi rimuoverlo dalla cima della pila. E' il metodo **LIFO** (*last in first out*).

Una coda (*queue*), invece è una struttura dati che implementa il metodo **FIFO** (*first in first out*). Immagina di entrare in una coda: la prima persona in coda è la prima a uscirne.

Ora approfondiremo ciascuno di questi concetti.

## Le pile

Uno stack è una struttura dati in cui gli elementi vengono aggiunti a un'estremità e un elemento viene rimosso solo da quella estremità. Le operazioni sono chiamate **push** (aggiunge un elemento alla fine dello stack) e **pop** (rimuove un elemento dalla fine dello stack).

Vediamo un semplice metodo per implementare uno stack in Python.

Inserisci ed esegui il seguente codice.

````python
stack = [] # creates empty list

def size_of_stack(stack):
    return len(stack)

# .append() functions identically here to push

stack.append(1)
print("Pushed one to stack:", stack, "and Current size of stack: ",size_of_stack(stack))

stack.append(2)
print("Pushed two to stack:", stack, "and Current size of stack: ",size_of_stack(stack))

stack.append(3)
print("Pushed three to stack:", stack, "and Current size of stack: ",size_of_stack(stack))

stack.pop()
print("Popped last item from stack", stack, "and Current size of stack: ",size_of_stack(stack))

stack.pop()
print("Popped last item from stack", stack, "and Current size of stack: ",size_of_stack(stack))

stack.pop()
print("Popped last item from stack", stack, "and Current size of stack: ",size_of_stack(stack))

stack.pop()
print("Popped last item from stack", stack, "and Current size of stack: ",size_of_stack(stack))
````

Osserviamo gli output della cella precedente. Quando si prova a estrarre un elemento da uno stack vuoto, viene generato un errore.

Esistono molti modi per implementare uno stack in Python, questo è solo uno dei modi più semplici.

## Le code

Vediamo un modo semplice per implementare le code in Python.

Per aggiungere un elemento alla fine di una coda e rimuoverne uno dalla finesi usano i metodi `enque` e `deque` .

Inserisci ed esegui il codice seguente per un'implementazione molto semplice di una coda in Python.

````python
queue = []

def size_of_queue(queue):
    return len(queue)

def enque(value):
    queue.append(value)

def deque():
    queue.pop(0)

enque(1)
print("Current queue: ", queue, "Size of queue:", size_of_queue(queue))

enque(2)
print("Current queue: ", queue, "Size of queue:", size_of_queue(queue))

enque(3)
print("Current queue: ", queue, "Size of queue:", size_of_queue(queue))

deque()
print("Current queue:", queue, "Size of queue:", size_of_queue(queue))

deque()
print("Current queue:", queue, "Size of queue:", size_of_queue(queue))

deque()
print("Current queue:", queue, "Size of queue:", size_of_queue(queue))
````

Come si vede, gli elementi aggiunti per primi sono anche i primi ad essere rimossi.

# Le eccezioni in Python
Da *go deploy lab*.

Vediamo cosa sono le eccezioni e come possono essere utilizzate per rilevare errori nel codice.

Un'eccezione rappresenta un errore. Il modo migliore per gestirla è con un blocco di codice **try...except**.

Impostiamo un'eccezione. Supponiamo di voler inserire l'altezza di una persona, ma sappiamo che non può essere negativa, quindi vogliamo che venga visualizzato un errore se viene inserita un'altezza non valida. Per esempio:

````python
height_in_cm = 180
if height_in_cm <= 0:
    raise Exception("Height cannot be less than or equal to zero CM!")
````

Questa sezione di codice non genera un'eccezione poiché l'altezza è > 0. Tuttavia, se modifichiamo `height_in_cm` a -10, ora si genera un'eccezione e viene visualizzato il testo specificato. Ad esempio, sostituiamo il codice precedente con il seguente:

````python
height_in_cm = -10
if height_in_cm <= 0:
    raise Exception("Height cannot be less than or equal to zero CM!")
````

Possiamo verificare determinate condizioni usando un'istruzione `assert`. Questa istruzione genera un errore se la condizione assertata restituisce `False`.

Ad esempio, possiamo asserire che l'altezza di input deve essere maggiore di 0. Quindi, se viene immessa un'altezza di -10 cm, il programma verrà interrotto. Per esempio:

````python
height_in_cm = -10
assert (height_in_cm >= 0), "Height cannot be negative!"
````

Possiamo usare i blocchi **try...except** per tentare di eseguire una determinata sezione di codice. Se quella sezione contiene un errore, l'intero corpo di codice non verrà bloccato. Per esempio:

````python
try:
    agaewrnwrn # clearly this is not correct code
except:
    print("Error") # this will return Error but won't halt the program
````

Inoltre, possiamo fare in modo che la sezione `except` stampi il tipo di errore. Per esempio:

````python
try:
    a = 1 / 0
except Exception as exception:
    print(exception)
````

1/0 è un errore di divisione per zero, quindi quando eseguiamo questo codice al di fuori della sezione *try except*, esso genera un errore e interrompe il codice; tuttavia, all'interno della sezione *try except*, esso stampa l'errore ma il programma continua.
